## Model Evaluation Metrics: Precision, Recall, and F1-Score

```python
print("\n--- Classification Metrics ---")

metrics = {
    'Model': [],
    'Metric': [],
    'Value': []
}

def evaluate_model(model_name, y_true, y_pred):
    global metrics
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"\n{model_name}:")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall: {recall:.4f}")
    print(f"  F1-Score: {f1:.4f}")

    metrics['Model'].extend([model_name, model_name, model_name])
    metrics['Metric'].extend(['Precision', 'Recall', 'F1-Score'])
    metrics['Value'].extend([precision, recall, f1])

# Evaluate Logistic Regression
evaluate_model('Logistic Regression', y_test, y_test_pred_lr)

# Evaluate Decision Tree
evaluate_model('Decision Tree', y_test, y_test_pred_dt)

# Evaluate Random Forest
evaluate_model('Random Forest', y_test, y_test_pred_rf)

df_metrics = pd.DataFrame(metrics)
display(df_metrics)
```

```python
# Reload the dataset to ensure a clean starting point for this subtask
df_raw = pd.read_csv(filename)

# Define features and target
X_raw = df_raw.drop(['Patient ID', 'Income', 'Heart Attack Risk'], axis=1)
y = df_raw['Heart Attack Risk']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_raw, y, test_size=0.2, random_state=42)

print("Libraries imported, dataset reloaded, features and target defined, and data split successfully.")
```

```python
class BloodPressureSplitter(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X_copy = X.copy()
        # Split 'Blood Pressure' into 'Systolic' and 'Diastolic'
        bp_split = X_copy['Blood Pressure'].str.split('/', expand=True)
        X_copy['Systolic'] = bp_split[0].astype(int)
        X_copy['Diastolic'] = bp_split[1].astype(int)

        # Drop the original 'Blood Pressure' column
        X_copy = X_copy.drop('Blood Pressure', axis=1)
        return X_copy

print("Custom BloodPressureSplitter transformer defined.")
```

**Reasoning**:
With the `BloodPressureSplitter` defined, I will apply it to a copy of `X_train` to identify the correct numerical and categorical features for the `ColumnTransformer`. This ensures that the newly created 'Systolic' and 'Diastolic' columns are included in the numerical features and that the original 'Blood Pressure' column is no longer present for categorical processing.



```python
X_train_transformed_bp = BloodPressureSplitter().fit_transform(X_train)

# Identify numerical features after BloodPressureSplitter transformation
numerical_features_for_scaling = X_train_transformed_bp.select_dtypes(include=['int64', 'float64']).columns.tolist()
# Remove 'Heart Attack Risk' if it somehow ends up here (it shouldn't, but as a safeguard)
numerical_features_for_scaling = [col for col in numerical_features_for_scaling if col != 'Heart Attack Risk']

# Identify categorical features after BloodPressureSplitter transformation
categorical_features_for_encoding = X_train_transformed_bp.select_dtypes(include=['object', 'category']).columns.tolist()

print(f"Numerical features for scaling: {numerical_features_for_scaling}")
print(f"Categorical features for encoding: {categorical_features_for_encoding}")
```

**Reasoning**:
Now that the numerical and categorical features have been identified, the next step is to instantiate the `StandardScaler` for numerical features and `OneHotEncoder` for categorical features, which will be used in the `ColumnTransformer` for preprocessing.



```python
# Create preprocessing transformers
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

print("Numerical and categorical transformers instantiated.")
```

**Reasoning**:
Now that the individual transformers are instantiated, the next logical step is to combine them into a `ColumnTransformer` (`preprocessor`). This `ColumnTransformer` will apply the `StandardScaler` to numerical features and the `OneHotEncoder` to categorical features, as identified previously.



```python
# Build the ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features_for_scaling),
        ('cat', categorical_transformer, categorical_features_for_encoding)
    ],
    remainder='drop'
)

print("ColumnTransformer (preprocessor) built successfully.")
```

**Reasoning**:
With the custom transformer and `ColumnTransformer` ready, the next step is to construct the complete machine learning pipeline, which will sequentially apply the `BloodPressureSplitter`, the `preprocessor`, and finally the `RandomForestClassifier`.



```python
# Construct the full machine learning pipeline
ml_pipeline = Pipeline(steps=[
    ('blood_pressure_splitter', BloodPressureSplitter()),
    ('preprocessor', preprocessor),
    ('randomforestclassifier', RandomForestClassifier(random_state=42))
])

print("Machine learning pipeline constructed successfully.")
```

**Reasoning**:
Now that the machine learning pipeline is constructed, the next step is to define the hyperparameter grid for the `RandomForestClassifier` within the pipeline, which will be used by `GridSearchCV` to explore different combinations of parameters.



```python
param_grid = {
    'randomforestclassifier__n_estimators': [50, 100, 200],
    'randomforestclassifier__max_depth': [5, 10, None],
    'randomforestclassifier__min_samples_split': [2, 5, 10]
}

print("Hyperparameter grid for RandomForestClassifier defined.")
```

**Reasoning**:
With the pipeline and parameter grid defined, the next step is to execute `GridSearchCV` to perform a comprehensive search for the optimal hyperparameters for the `RandomForestClassifier` within the defined pipeline.



```python
# Instantiate GridSearchCV
grid_search = GridSearchCV(
    estimator=ml_pipeline,
    param_grid=param_grid,
    cv=3,  # Using 3-fold cross-validation
    scoring='f1', # Using F1-score as the evaluation metric
    n_jobs=-1 # Use all available cores
)

# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print(f"Best parameters found: {grid_search.best_params_}")
print(f"Best F1-score found: {grid_search.best_score_:.4f}")
```

**Reasoning**:
After running `GridSearchCV` and identifying the best parameters, the final step in the instructions is to extract the best performing model (pipeline) from the `grid_search` object and store it in a variable named `best_pipeline`.



```python
best_pipeline = grid_search.best_estimator_

print("Best performing pipeline extracted successfully.")
```
